```python
# Convert a list of numbers [1, 4, 9, 16] into their square roots.
def sq_roots(num):
    s=num/2
    for i in range(9):
        s=(s+num/s)/2
    return s
List=[1, 4, 9, 16]
print(list(map(lambda x:int(sq_roots(x)),List)))

```

```python
# D. List comprehension vs map
# Rewrite Above questions using a list comprehension.

# Given [1, 2, 3, 4], use map to get [2, 4, 6, 8].

# Convert a list of numbers [1, 4, 9, 16] into their square roots.

# Given a list of names, convert all names to upper
```

```python
# LEVEL 3 — EASY–MEDIUM (introducing filter)
# E. filter() only                                                              filter(function, iterable)


# From [1, 2, 3, 4, 5, 6], keep only even numbers.

# From [10, 15, 20, 25, 30], keep numbers greater than 20.

# From a list of strings, keep only words longer than 4 characters.

# function: A callable (like a named function or a lambda function) that accepts one argument and returns a boolean value (True or False). If None is passed, the filter removes all "falsy" values (e.g., 0, "", False, None).
# iterable: Any sequence to be filtered, such as a list, tuple, set, or string. 
```

```python
# filter(function, iterable)

# From [1, 2, 3, 4, 5, 6], keep only even numbers.

List=[1, 2, 3, 4, 5, 6]
list(filter(lambda x:x%2==0, List))
# list(filter(lambda x:True if x%2==0 else False, List))

```

```python
# From [10, 15, 20, 25, 30], keep numbers greater than 20.

List=[10, 15, 20, 25, 30]
list(filter(lambda x: x>20,List))
```

```python
# From a list of strings, keep only words longer than 4 characters

List=["hello","land","robot","cat","pneumonoultramicroscopicsilicovolcanoconiosis","dog"]
list(filter(lambda x:len(x)>4,List))
```

```python
# Rewrite above questions using list comprehension

# From [1, 2, 3, 4, 5, 6], keep only even numbers.
List=[1, 2, 3, 4, 5, 6]
print([x for x in List if x%2==0])

# From a list of strings, keep only words longer than 4 characters.
List=["hello","land","robot","cat","pneumonoultramicroscopicsilicovolcanoconiosis","dog"]
print([x for x in List if len(x)>4])
```

```python
# LEVEL 4 — MEDIUM (introducing reduce)
# G. reduce() basics                                reduce(function, iterable, [initializer])


# (Use from functools import reduce)

# Find the sum of [1, 2, 3, 4] using reduce.

# Find the product of [1, 2, 3, 4] using reduce.

# Find the maximum number in [3, 7, 2, 9] using reduce.

# function: A function that takes exactly two arguments and returns a single value. This function is applied cumulatively to the elements of the iterable.
# iterable: A sequence (list, tuple, etc.) whose elements are to be reduced.
# initializer (optional): An initial value that is placed before the first item of the iterable in the calculation. 
# If the iterable is empty and no initializer is provided, a TypeError is raised. 
```

```python
# Find the sum of [1, 2, 3, 4] using reduce.
from functools import reduce
List=[1, 2, 3, 4]
reduce(lambda x,y: x+y ,List)

reduce(lambda x,y:x*y,List)

# c=0
# for i in List:
#     c+=i
# print(c)
```

```python
# Find the maximum number in [3, 7, 2, 9] using reduce.
List=[3, 7, 2, 9]
reduce(lambda x,y: x if x>y else y,List)            # equivalent to max()
```

```python
# LEVEL 5 — MEDIUM (combining map + filter)
# H. Map + Filter

# From [1, 2, 3, 4, 5, 6], keep even numbers and then square them.

# From a list of numbers, keep numbers divisible by 3 and multiply them by 10.

# From a list of names, keep names starting with "A" and convert them to lowercase.
```

```python
# From [1, 2, 3, 4, 5, 6], keep even numbers and then square them.  Map + Filter
List=[1, 2, 3, 4, 5, 6]
list(map(lambda x: x**2,filter(lambda x: x%2==0,List)))         #first filter the list into even numbers only list then use map function to sq the numbers
#filter return based on True
```

```python
# From a list of numbers, keep numbers divisible by 3 and multiply them by 10.
List=[1,2,3,4,5,6,7,8,9,12,24]
list(map(lambda x : x*10,filter(lambda x:x%3==0,List)))
```

```python
# From a list of names, keep names starting with "A" and convert them to lowercase.
List=["Arnold","AJ","David","John","Castiel"]
list(map(lambda x:x.lower(),filter(lambda x:x[0]=="A",List)))
```

```python
# LEVEL 8 — MEDIUM (filter → map → reduce pipeline)
# K. Full pipeline (no tricks)

    # From [5, 10, 15, 20]:

        # keep values ≥ 10

        # double them

        # find the total

    # From a list of salaries:

        # remove salaries below 30,000

        # add 10% increment

        # calculate total payroll
```

```python
  # From [5, 10, 15, 20]:

        # keep values ≥ 10

        # double them

        # find the total

List=[5, 10, 15, 20]
sum([x*x for x in List if x>=10])

# using filter → map → reduce pipeline

List=[5, 10, 15, 20]
reduce(lambda x,y:x+y,map(lambda x:x*x,filter(lambda x:x>=10,List)))  
# Note if you try to list() it will throw error Python gets confused 
# because you can't turn a single number into a list—you can 
# only turn collections (like sets or tuples) into lists.

#in short
# The filter (The Gatekeeper)
# The map (The Transformer)
# The reduce (The Crusher)

```

```python
   # From a list of salaries:

        # remove salaries below 30,000

        # add 10% increment

        # calculate total payroll

salary=[10000,30000,40000,50000,20000,5000]
sum([x+(x*0.1) for x in salary if x<30000])

list(filter(lambda x:x<30000,salary))
reduce(lambda x,y:x+y,map(lambda x:x*1.1,filter(lambda x:x<30000,salary)))
```

```python
# iterable (required): The sequence or collection you want to sort.

# key (optional): A function to be called on each element to customize the sort order (e.g., key=len to sort by length, or a lambda function).

# reverse (optional): A Boolean value. If True, the list is sorted in descending order. The default is False (ascending order)

#  sorted(iterable, key=None, reverse=False)
```

```python
# Sort on basis of second item of tuple                     sorted(iterable, key=None, reverse=False)

List=[(1, 10), (2, 5), (3, 20)]
sorted(List,key=lambda x: x[1])
sorted(List,key=lambda x: x[1],reverse=True)
```

```python
# Sort the list by price in ascending order                  sorted(iterable, key=None, reverse=False)
# Sort the list by price in descending order
products = [("Pen", 10), ("Notebook", 50), ("Pencil", 5), ("Eraser", 7)]
print(sorted(products,key=lambda x:x[1]))
print(sorted(products,key=lambda x:x[1],reverse=True))
```
