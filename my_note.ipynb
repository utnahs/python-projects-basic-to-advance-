```python
# to chech range in iterables
list=["a","b","c","d","e","f","g","h"]
print(list)
sliced=list[0:2]        #[starts from the first, counts from the first i.e "1"]
sli=list[0:7:2] #start:stop:gap here skips 1
print(sliced,sli)
print(list[3:5]) #starts from 3rd("c" as python indexes from 0), counts from 1 ends at 5th position

```

```python

tuple=(45,46,47,48,49)
print(tuple)
sliced_=tuple[0:2]
print(sliced_)
```

```python
# “Dictionaries don’t support slicing because they are key-based, not position-based.”
dict={"Name":"Shantu","Age":27,"Status":"Unmarried"}
print(dict)
print(dict["Name"])
print(dict["Name"], dict["Age"])

```

```python
# negative slicing STRINGS
#always prints from left to right be it positve slicing or negative
s="hello"
print(s[-1])
#print(s[-10]) will show an error index out of range
print(s[:])
print(s[-4:-1])
'''print(s[1:2]) #prints the single character 
print(s[-1:2]) #doesnt return any thing
print(s[-1:-3]) # can't print in reverse'''
print(s[::-1]) #prints the reverse of string, or any iterables
```

```python
l=[[5,6,7,8],[1,2,3,4]]
for i in l:  ### this line creates an iterator thus python knows which value is next
    for j in i[::-1]: 
        print(j)

    
```
